```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score
```

```python
df = df.astype('category')
df.dtypes
```

```python
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])
```

```python
df.head()
```

```python
df = df.drop(["veil-type"],axis=1)
```

```python
df.corr()
```

```python
X = df.drop("class", axis=1)
y = df["class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)
```

```python
sv = SVC(gamma="auto")
sv.fit(X_train, y_train)
```

```python
sv.score(X_test, y_test)
```

```python
y_predict = sv.predict(X_test)
```

```python
y_predict = sv.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

Predicting and estimating the result

```python
y_pred_dt = dt.predict(X_test)
print("Decision Tree Classifier report: \n",classification_report(y_test,y_pred_dt))
print("Test Accuracy: {}%".format(round(dt.score(X_test, y_test)*100,2)))
```

```python
from sklearn.linear_model import LogisticRegression
l = LogisticRegression(solver="liblinear", max_iter=500)
l.fit(X_train, y_train)
print("Test Accuracy: {}%".format(round(l.score(X_test, y_test)*100,2)))
```

```python
y_predict_l = l.predict(X_test)
print("Logistic Regression Classifier report:\n\n", classification_report(y_test, y_predict_l))
```

```python
l.score(X_test, y_test)
```

```python
l.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score, roc_curve
y_pred= l.predict(X_test)
y_true = y_test
cm = confusion_matrix(y_true, y_pred)
f, ax = plt.subplots(figsize =(8,7))
sns.heatmap(cm,annot = True,linewidths=0.5,linecolor="blue",fmt = ".0f",ax=ax)
plt.xlabel("y_predict")
plt.ylabel("y_true")
plt.show()
```

```python
kv = sorted([(k, v) for k, v in zip(X_train.columns, dt.feature_importances_)], key=lambda x: x[1], reverse=True)
keys, values = [x[0] for x in kv], [x[1] for x in kv]

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(keys, values)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:ax.spines[s].set_visible(False)
    
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',linestyle ='-.', linewidth = 0.5,alpha = 0.3)
 
# Show top values
ax.invert_yaxis()
 
# Add Plot Title
ax.set_title('Feature Importance',loc ='left', )
 
# Show Plot
```
